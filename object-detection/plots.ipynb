```python
import matplotlib.pyplot as plt
```

```python
# Quantization configurations
quant_configs = ['int8', 'float16', 'float32', 'base']

# Lists to store results
AP = [0.168, 0.336, 0.335, 0.414]
AR = [0.377,0.582,0.582,0.657]
times = [205,187,184,215]
sizes = [7.7, 12.9, 24.9, 33.4]
wH = [2.75, 2.476, 2.46, 2.90]
```

```python
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()
fig3, ax3 = plt.subplots()
fig4, ax4 = plt.subplots()
fig5, ax5 = plt.subplots()

# Plotting Accuracy
ax1.plot(quant_configs, AP, marker='o', label='Average Precision')
ax1.set_title('Precision vs Quantization Configuration')
ax1.set_xlabel('Quantization Configuration')
ax1.set_ylabel('Average Precision')
ax1.legend()

# Plotting Time
ax2.plot(quant_configs, times, marker='o', label='Time')
ax2.set_title('Interence time vs Quantization Configuration')
ax2.set_xlabel('Quantization Configuration')
ax2.set_ylabel('Time (s)')
ax2.legend()

# Plotting Size
ax3.plot(quant_configs, sizes, marker='o', label='Model Size')
ax3.set_title('Model Size vs Quantization Configuration')
ax3.set_xlabel('Quantization Configuration')
ax3.set_ylabel('Size (MB)')
ax3.legend()

# Plotting wH
ax4.plot(quant_configs, wH, marker='o', label='wH')
ax4.set_title('watt/h vs Quantization Configuration')
ax4.set_xlabel('Quantization Configuration')
ax4.set_ylabel('wH')
ax4.legend()

# Plotting AR 
ax5.plot(quant_configs, AR, marker='o', label='Average Recall')
ax5.set_title('Recall vs Quantization Configuration')
ax5.set_xlabel('Quantization Configuration')
ax5.set_ylabel('Average Recall')


plt.show()
```
